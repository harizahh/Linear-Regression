```python
# importing the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('linear_reg.xlsx',sheet_name=1,header=3,nrows=27,usecols=[1,2,3,4,5])
df.rename(columns={'Number of weekly riders':'num_riders','Price per week':'price','Population of city':'population',
                  'Monthly income of riders':'income','Average parking rates per month':'avg_parking'},inplace=True)
df.sample()
```

```python
# Identifying the features and Target datas
x=df.drop(['num_riders'],axis=1)
y=df['num_riders']
```

```python
# visualization
plt.figure(figsize=(15,5))
plt.subplot(2,2,1)
plt.scatter('price','num_riders',data=df,color='r')

plt.subplot(2,2,2)
plt.scatter('population','num_riders',data=df,color='b')

plt.subplot(2,2,3)
plt.scatter('income','num_riders',data=df,color='g')

plt.subplot(2,2,4)
plt.scatter('avg_parking','num_riders',data=df,color='y')

plt.show()
```

```python
sns.boxplot(df)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(2,3,1)
sns.boxplot(y='num_riders',data=df,color='r')

plt.subplot(2,3,2)
sns.boxplot(y='price',data=df,color='g')

plt.subplot(2,3,3)
sns.boxplot(y='population',data=df,color='b')

plt.subplot(2,3,4)
sns.boxplot(y='income',data=df,color='y')

plt.subplot(2,3,5)
sns.boxplot(y='avg_parking',data=df,color='r')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2).fit(x,y)
poly_x=poly.transform(x)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(poly_x,y)
```

```python
y_pred=model.predict(poly_x)
```

```python
model.score(poly_x,y)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,y_pred)
```
